```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = data[data['Income'].notnull()]
data.reset_index(drop=True, inplace=True)
data
```

```python
data
```

```python
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
data
```

```python
data.describe(exclude='number')
```

```python
data['Year'] = data['Dt_Customer'].apply(lambda row: row.year)
data
```

```python
data[['Year']].describe()
```

```python
data['Age'] = data['Year'] - data['Year_Birth']
data
```

```python
data1 = data.copy()
data1
```

```python
data1.drop([
    'ID',
    'Year_Birth',
    'Education',
    'Marital_Status',
    'Dt_Customer',
    'Year'
],axis=1, inplace=True)  
```

```python
data1.rename({
    'MntWines': 'WineRank',
    'MntFruits': 'FruitsRank',
    'MntMeatProducts': 'MeatRank',
    'MntFishProducts': 'FishRank',
    'MntSweetProducts': 'SweetRank',
    'MntGoldProds': 'GoldRank',
    'NumWebPurchases': 'WebRank',
    'NumCatalogPurchases': 'CatalogRank',
    'NumStorePurchases': 'StoreRank'
}, axis=1, inplace=True)

data1
```

```python
outlier_age = data1.loc[data1['Age'] > 90].index
data1.drop(outlier_age, inplace=True)
data1.reset_index(drop=True, inplace=True)

sns.kdeplot(data1['Age'])
```

```python
data1['Kidhome'] = data1['Kidhome'].apply(lambda row: 1 if row >= 1 else 0)
data1['Teenhome'] = data1['Teenhome'].apply(lambda row: 1 if row >= 1 else 0)

data1
```

```python
from sklearn.manifold import TSNE, LocallyLinearEmbedding, MDS, Isomap
from sklearn.preprocessing import StandardScaler

data2 = data1.copy()
data2
```

```python
data2['Accepted'] = data2[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].sum(axis=1) > 0
data2['Accepted'] = data2['Accepted'].apply(lambda row: 1 if row else 0)

data2.drop(['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response'], axis=1, inplace=True)

data2
```

```python
data2 = pd.DataFrame(StandardScaler().fit_transform(data2), columns=data2.columns)
```

```python
for i in data2.columns:
    data2[i].astype(dtype=float)

data2.info()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)

pca_data = pd.DataFrame(pca.fit_transform(data2))

plt.plot(pca.explained_variance_ratio_.cumsum())

data2 = pd.concat([data2, pca_data], axis=1)

data2
```
